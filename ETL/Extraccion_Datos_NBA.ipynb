```python
import os
import zipfile
import sqlite3
import pandas as pd
from google.colab import drive

# 1. Definir la ruta exacta que acabamos de descubrir
ruta_base = "/content/drive/MyDrive/Proyecto Final - Grupo 7 - DAPT 10 - Data Analytics"
archivo_zip = os.path.join(ruta_base, "nba.sqlite.zip")
carpeta_destino = "/content/nba_descomprimido"

# 2. Descomprimir el archivo .zip
print("‚è≥ Descomprimiendo la base de datos...")
with zipfile.ZipFile(archivo_zip, 'r') as zip_ref:
    zip_ref.extractall(carpeta_destino)
print("‚úÖ Descompresi√≥n completada.")

# 3. Localizar el archivo .sqlite dentro de la carpeta descomprimida
# A veces al descomprimir se crea otra carpeta adentro, vamos a buscar el archivo .sqlite
for root, dirs, files in os.walk(carpeta_destino):
    for file in files:
        if file.endswith(".sqlite"):
            path_sqlite_local = os.path.join(root, file)
            print(f"üìç Base de datos encontrada en: {path_sqlite_local}")

# 4. Conectar y extraer datos de prueba
conn = sqlite3.connect(path_sqlite_local)
df_players = pd.read_sql_query("SELECT person_id, display_first_last, position FROM common_player_info LIMIT 10", conn)
conn.close()

print("\nüìä Primeros 10 jugadores extra√≠dos con √©xito:")
display(df_players)
```
