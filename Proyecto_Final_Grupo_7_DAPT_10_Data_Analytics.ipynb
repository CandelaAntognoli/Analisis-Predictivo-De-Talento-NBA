```python
# Limpieza y normalizacion
import numpy as np
import pandas as pd

df_player = pd.read_sql("""
SELECT
  person_id,
  display_first_last,
  position,
  height,
  weight,
  country,
  team_id,
  from_year,
  to_year
FROM common_player_info
""", conn)

def height_to_cm(h):
    try:
        feet, inches = str(h).split("-")
        return (int(feet) * 12 + int(inches)) * 2.54
    except:
        return np.nan

df_player["height_cm"] = df_player["height"].apply(height_to_cm)
df_player["weight_lbs"] = pd.to_numeric(df_player["weight"], errors="coerce")
df_player["weight_kg"] = df_player["weight_lbs"] * 0.453592

df_player["position"] = df_player["position"].replace(["", "None", "nan"], np.nan)
df_player["country"] = df_player["country"].replace(["", "None", "nan"], np.nan)

dim_player = (
    df_player[["person_id","display_first_last","position","country","height_cm","weight_kg"]]
    .rename(columns={"display_first_last":"player_name"})
    .drop_duplicates(subset=["person_id"])
    .reset_index(drop=True)
)

print("‚úÖ DIM_PLAYER:", dim_player.shape)
dim_player.head()
```

```python
# Creamos la columna team_id
dim_team = pd.read_sql("""
SELECT
  id AS team_id,
  abbreviation,
  nickname,
  city,
  year_founded
FROM team
""", conn)

dim_team["team_id"] = pd.to_numeric(dim_team["team_id"], errors="coerce").astype("Int64")
dim_team["year_founded"] = pd.to_numeric(dim_team["year_founded"], errors="coerce").astype("Int64")

dim_team = dim_team.dropna(subset=["team_id"]).drop_duplicates(subset=["team_id"]).reset_index(drop=True)

print("‚úÖ DIM_TEAM:", dim_team.shape)
dim_team.head()
```

```python
# Normalizamos la tabla DIM_DATE
df_game_dates = pd.read_sql("SELECT DISTINCT game_date FROM game", conn)
df_game_dates["full_date"] = pd.to_datetime(df_game_dates["game_date"], errors="coerce")
df_game_dates = df_game_dates.dropna(subset=["full_date"])

dim_date = pd.DataFrame({
    "date_key": df_game_dates["full_date"].dt.strftime("%Y%m%d").astype(int),
    "full_date": df_game_dates["full_date"].dt.date,
    "year": df_game_dates["full_date"].dt.year,
    "month": df_game_dates["full_date"].dt.month,
    "day": df_game_dates["full_date"].dt.day,
    "season_year": df_game_dates["full_date"].dt.year
}).drop_duplicates(subset=["date_key"]).reset_index(drop=True)

print("‚úÖ DIM_DATE:", dim_date.shape)
dim_date.head()
```

```python
# Normalizamos la columna MAP_PLAYER_TEAM
map_player_team = df_player[["person_id","team_id","from_year","to_year"]].copy()

map_player_team["team_id"] = pd.to_numeric(map_player_team["team_id"], errors="coerce")
map_player_team["from_year"] = pd.to_numeric(map_player_team["from_year"], errors="coerce")
map_player_team["to_year"] = pd.to_numeric(map_player_team["to_year"], errors="coerce")

map_player_team = map_player_team.dropna(subset=["team_id","from_year","to_year"])
map_player_team["team_id"] = map_player_team["team_id"].astype(int)
map_player_team["from_year"] = map_player_team["from_year"].astype(int)
map_player_team["to_year"] = map_player_team["to_year"].astype(int)

map_player_team = map_player_team.drop_duplicates(subset=["person_id","team_id","from_year"]).reset_index(drop=True)

print("‚úÖ MAP_PLAYER_TEAM:", map_player_team.shape)
map_player_team.head()
```

```python
# Normalizamos la tabla FACT_COMBINE
combine = pd.read_sql("""
SELECT
  season,
  player_id,
  max_vertical_leap,
  lane_agility_time,
  three_quarter_sprint,
  wingspan,
  bench_press
FROM draft_combine_stats
""", conn)

combine["season_year"] = pd.to_numeric(combine["season"].astype(str).str[:4], errors="coerce")

for c in ["max_vertical_leap","lane_agility_time","three_quarter_sprint","wingspan","bench_press"]:
    combine[c] = pd.to_numeric(combine[c], errors="coerce")

fact_combine = (
    combine.rename(columns={"player_id":"person_id"})
    [["person_id","season_year","max_vertical_leap","lane_agility_time","three_quarter_sprint","wingspan","bench_press"]]
    .dropna(subset=["person_id","season_year"])
    .reset_index(drop=True)
)

print("‚úÖ FACT_COMBINE:", fact_combine.shape)
fact_combine.head()
```

```python
# Construimos FACT_GAME desde Games + limpieza
import pandas as pd
import numpy as np

df_game = pd.read_sql("""
SELECT
  game_id,
  game_date,
  team_id_home,
  team_id_away,
  pts_home,
  ast_home,
  tov_home,
  blk_home,
  stl_home,
  plus_minus_home,
  pts_away,
  ast_away,
  tov_away,
  blk_away,
  stl_away,
  plus_minus_away
FROM game
""", conn)

# --- Normalizaci√≥n de tipos ---
# ids de equipos a num√©rico
df_game["team_id_home"] = pd.to_numeric(df_game["team_id_home"], errors="coerce")
df_game["team_id_away"] = pd.to_numeric(df_game["team_id_away"], errors="coerce")

# fecha -> date_key
df_game["full_date"] = pd.to_datetime(df_game["game_date"], errors="coerce")
df_game["date_key"] = df_game["full_date"].dt.strftime("%Y%m%d")

# m√©tricas num√©ricas (por si vienen como object)
metric_cols = [
    "pts_home","ast_home","tov_home","blk_home","stl_home","plus_minus_home",
    "pts_away","ast_away","tov_away","blk_away","stl_away","plus_minus_away"
]
for c in metric_cols:
    df_game[c] = pd.to_numeric(df_game[c], errors="coerce")

# --- Selecci√≥n final (DW staging) ---
fact_game = df_game[[
    "game_id",
    "date_key",
    "team_id_home",
    "team_id_away",
    "pts_home","ast_home","tov_home","blk_home","stl_home","plus_minus_home",
    "pts_away","ast_away","tov_away","blk_away","stl_away","plus_minus_away"
]].dropna(subset=["game_id","date_key","team_id_home","team_id_away"])

# tipos finales
fact_game["date_key"] = pd.to_numeric(fact_game["date_key"], errors="coerce").astype("Int64")
fact_game["team_id_home"] = fact_game["team_id_home"].astype("Int64")
fact_game["team_id_away"] = fact_game["team_id_away"].astype("Int64")

fact_game = fact_game.drop_duplicates(subset=["game_id"]).reset_index(drop=True)

print("‚úÖ FACT_GAME:", fact_game.shape)
fact_game.head()
```
